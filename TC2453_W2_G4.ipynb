## Data Acquistion

Step 1: Start by importing the neccesary library and loads the dataset from local system

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from google.colab import files
```

Step 2: Choose the csv file to upload from your file system (Choose the hour.csv )


```python
# function to upload/mount datasets from local
csv_file_name =''

_files = files.upload()
if len(_files) >0:
    for k,v in _files.items():
       open(k,'wb').write(v)
       csv_file_name = k
   
```

```python
# read CSV file using pandas
data = pd.read_csv(csv_file_name)
data.head()
```

## Data Preprocessing

Checks for any null values in the Dataset

```python
# check for any null values in the dataset
print("Data contains null/invalid values: ", data.isnull().values.any())
```

Select a few features to be evaluate from the dataset

```python
data = data[["season","hr","weathersit","cnt"]]
```

```python
# This is a function to plot
def plotScatter(x,y):
  plt.scatter(data[x],y,color="green")
  plt.xlabel(x)
  plt.ylabel("Bike rental")
  plt.show()

# Count of Bike Rentals
y = data["cnt"]
for col in data.columns[:-1]:
  plotScatter(col,y)


# We will define a function to take a single independent feature as input and 
# output the results
#cols = ['weathersit', 'season', 'hr']

```
